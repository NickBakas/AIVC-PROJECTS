```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import seaborn as sns
from sklearn.model_selection import cross_val_score, KFold
from sklearn.model_selection import train_test_split

# Read the Excel file
data1 = pd.read_excel('D_PreppedLogFile (1).xlsx')
data2=pd.read_csv('B_TravelDataUsers.csv')


# Display the first few rows of the DataFrame
data1.head()


# Read the first sheet into a DataFrame
maindata= pd.read_excel('D_PreppedLogFile (1).xlsx', sheet_name='LogFile Prepped and Pivoted')

columns_to_convert = maindata.loc[:, 'Touchpoint 1':'Touchpoint 19'].columns

# Iterate over each column and apply extraction and conversion
for column in columns_to_convert:
    maindata[column] = maindata[column].fillna(0)  # Replace NaN with -1 or any other placeholder value
    maindata[column] = maindata[column].astype(str).str.extract('(\d+)').astype(int)

maindata.head()
```

```python
columns_to_extract = maindata.iloc[:, 13:32]  # Adjust the column indices as needed

user_id_column = maindata['UserID']

arrays_with_user_id = pd.concat([user_id_column, columns_to_extract], axis=1)

seqdf = pd.DataFrame({'UserID': arrays_with_user_id['UserID'], 'Touchpoint Sequences': arrays_with_user_id.values[:, 1:].tolist()})
seqdf.head()
```

```python
flattened_sequences = seqdf['Touchpoint Sequences'].explode()

mean_by_row = flattened_sequences.groupby(level=0).mean()
std_by_row = flattened_sequences.groupby(level=0).std()
```

```python
non_zero_sequences = seqdf['Touchpoint Sequences'].apply(lambda x: [val for val in x if val != 0])
# Plot histogram of non-zero sequence lengths
plt.hist(non_zero_sequences.apply(len), bins=20)
plt.xlabel('Sequence Length')
plt.ylabel('Frequency')
plt.title('Distribution of Non-zero Sequence Lengths')
plt.show()
```

```python
def count_touchpoints(sequence):
    touchpoint_counts = {}
    for touchpoint in sequence:
        if touchpoint != 0:  # Exclude zero-padding
            if touchpoint in touchpoint_counts:
                touchpoint_counts[touchpoint] += 1
            else:
                touchpoint_counts[touchpoint] = 1
    return touchpoint_counts

# Apply the function to each sequence in the DataFrame
touchpoint_counts_per_row = seqdf['Touchpoint Sequences'].apply(count_touchpoints)

# Convert the dictionary of counts into a DataFrame
touchpoint_counts_df = pd.DataFrame(touchpoint_counts_per_row.tolist())

# Fill missing values (i.e., touchpoints that didn't appear in a sequence) with zeros
touchpoint_counts_df.fillna(0, inplace=True)

# Display the resulting DataFrame
print(touchpoint_counts_df)

touchpoint_counts_df.sum().plot(kind='bar')
plt.xlabel('Touchpoint')
plt.ylabel('Count')
plt.title('Count of Touchpoints')
plt.show()
```
